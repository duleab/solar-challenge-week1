```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load cleaned data

```python
# Load the datasets
benin = pd.read_csv(r'..\data\benin_clean.csv')
sierra_leone = pd.read_csv(r'..\data\sierraleone_clean.csv')
togo = pd.read_csv(r'..\data\togo_clean.csv')
# Add country labels
benin["Country"] = "Benin"
sierra_leone["Country"] = "Sierra Leone"
togo["Country"] = "Togo"
```

```python

# Combine datasets
df_all = pd.concat([benin, sierra_leone, togo], ignore_index=True)
```

###  Boxplots for GHI, DNI, DHI

```python
metrics = ["GHI", "DNI", "DHI"]
for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df_all, x="Country", y=metric, palette="Set2")
    plt.title(f"{metric} Distribution by Country")
    plt.ylabel(f"{metric} (W/m)")
    plt.grid(True, axis='y')
    plt.show()
```

###  Summary Table: mean, median, std

```python
summary = df_all.groupby("Country")[["GHI", "DNI", "DHI"]].agg(["mean", "median", "std"]).round(2)
summary
```

#### Bar Chart of Avg GHI


```python
avg_ghi = df_all.groupby("Country")["GHI"].mean().sort_values(ascending=False)
avg_ghi.plot(kind="bar", color="skyblue", figsize=(6, 4))
plt.title("Average GHI by Country")
plt.ylabel("GHI (W/m)")
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.show()
```
